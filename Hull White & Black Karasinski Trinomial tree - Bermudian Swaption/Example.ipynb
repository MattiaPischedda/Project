#### Parameters setting
- **note:** Given the yield curve provided, the process has been modeled with a time step dt equal to the one of the term structure

```python
yield_curve = [(0.5*i, 0.03+0.001*i) for i in range(1, 10)]


sim1  = black_Karasinski(0.06, 0.08, 4, 8 , yield_curve)
sim2 = hull_white(0.25, 0.01, 1, 12 , r)
sim2.print_tree()

```
